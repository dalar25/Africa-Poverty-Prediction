```python
def match_data(folder_dir,countryName,countryExceptions):
    
    df = []

    for imageName in os.listdir(folder_dir):
        if (imageName.endswith(".png")):
            path = os.path.join(folder_dir, imageName)

            Latitude = imageName.split("_")[0]
            Longitude = imageName.split("_")[1]
            try:
                location = geolocator.reverse(Latitude+","+Longitude)
            except:
                continue
            location  [i.strip() for i in str(location).split(",")]
            country = location[len(location)-1]
            countryException = False
            for i in countryExceptions:
                if country == i: countryException = True
            if countryException:
                country = countryName
            
            if country != countryName:
                Latitude = imageName.split("_")[2]
                Longitude = imageName.split("_")[3].replace('.png','')
                location = geolocator.reverse(Latitude+","+Longitude)
                location = [i.strip() for i in str(location).split(",")]
                country = location[len(location)-1]
                countryException = False
                for i in countryExceptions:
                    if country == i: countryException = True
                if countryException:
                    country = countryName

            econVal = 0
            if country == countryName:
                for part in location:
                    try:
                        econVal = data[country][0][data[country][1][part]]
                    except:
                        pass
            
            inDict = (econVal != 0)
            if not(inDict): 
                for i in location:
                    print(i)
                print()
            else:
                df.append([path,econVal])

    df = pd.DataFrame(df, columns=['Image', 'PPP'])

    return df
```
